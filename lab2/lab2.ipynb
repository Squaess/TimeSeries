```python
def my_fft_plot(t, x, N):
    delta = t[2] - t[1]
    xfft = scipy.fftpack.fft(x)
    freq = np.linspace(0, 1./(2*delta), N//2)
    plt.plot(freq, 2*np.abs(xfft[:N//2]/N))
```
