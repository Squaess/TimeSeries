```python
import pywt
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 1024
t = np.linspace(0, 3, N)
f = 5
a = pywt.central_frequency('mexh')/(delta*f)
scale = 17
x = np.sin(2 * np.pi * a * t)
plt.plot(t, x)
```

```python
delta = t[2] - t[1]
coef, freqs = pywt.cwt(x, scale, 'mexh', delta)

print(freqs)
print(coef)
```

```python
fig, (ax1, ax2) = plt.subplots(2,1)
ax1.plot(t, x)
ax1.set_title('sinusoid')
ax2.plot(t, np.abs(coef[0][:]))
ax2.set_title('cwt')

plt.show()
```
